Parte 1: Preparaci√≥n 

```python
data["Gender"]=pd.Series(np.searchsorted(["Male","Female"], data.Gender.values), data.index)
data.head(3)
```

```python
data.tail(3)
```

```python
data["Married"]=pd.Series(np.searchsorted(["Yes","No"], data.Married.values), data.index)
```

```python
data.head()
```

```python
data["Student"]=pd.Series(np.searchsorted(["Yes","No"], data.Student.values), data.index)
```

```python
data.head()
```

```python
df=pd.get_dummies(data.Ethnicity)
df
```

```python
df.drop("African American", axis="columns", inplace=True)
df
```

```python
df2=pd.concat([data,df], axis=1)
df2
```

## Parte 2: Ajustar el modelo

```python
y=df2["Balance"]
y
```

```python
X=df2.drop(["Balance","Ethnicity"], axis="columns")
X
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, shuffle=True, test_size=0.25)
```

```python
model=LinearRegression()
regressor=model.fit(X_train, y_train)
```

```python
yhat=regressor.predict(X_test)
```

```python
r2_score(y_test,yhat)
```

## Parte 3: Clasificador y regresor

```python
part3=df2; part3
```

```python
part3["Is_Zero"]=np.where(part3["Balance"]==0, True, False)
part3.drop(["Ethnicity"], axis="columns", inplace=True)
part3
```

```python
y_knn=part3["Is_Zero"]
X_knn=part3.drop("Is_Zero", axis="columns")
X_knn_train, X_knn_test, y_knn_train, y_knn_test= train_test_split(X_knn, y_knn, test_size=0.25, shuffle=True)
```

```python
y_knn
```

```python
#Knn tiene 1 hyperparametro
parameters={"n_neighbors":range(1,100)}
knn_model=KNeighborsClassifier()
gs=GridSearchCV(knn_model, parameters, cv=5, n_jobs=1, verbose=1, return_train_score=True)
gs.fit(X_knn_train, y_knn_train)
```

```python
gs.best_estimator_
```

```python
gs.score(X_knn_test, y_knn_test)
```

```python
yhat_knn=gs.predict(X_knn_test)
pd.Series(yhat_knn).value_counts()
```

```python
select=pd.Series(y_knn).map(lambda x: True if x==False else False)
final_df=part3[select].drop("Is_Zero", axis=1)
final_df
```

```python
y_final=final_df["Balance"]
```

```python
X_final=final_df.drop("Balance", axis=1,)
X_final
```

```python
X_train, X_test, y_train, y_test= train_test_split(X_final, y_final, test_size=0.25)
```

```python
linear_model=LinearRegression()
linear_regressor=linear_model.fit(X_train, y_train)
```

```python
ypred=linear_model.predict(X_test)
```

```python
r2_score(y_test,ypred)
```

```python
fig, ax=plt.subplots(figsize=(15,8))
ax.scatter(y_test,ypred, color="cyan", marker="+")
ax.set_title("Y verdaderovs Y calculado")
ax.set_ylabel ("y real")
ax.set_xlabel ("y calculado")
```
